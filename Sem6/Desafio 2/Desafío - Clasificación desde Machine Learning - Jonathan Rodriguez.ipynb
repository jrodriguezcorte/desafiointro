```python
df = df_original.copy()
df['student_yes'] = pd.get_dummies(df['student'], drop_first=True)
df['default_yes'] = pd.get_dummies(df['default'], drop_first=True)
df = df.drop(columns='student')
df = df.drop(columns='default')
df.head()
```

```python
sns.heatmap(df.corr(), cmap='Blues', annot=True);
```

```python
m1_logit = smf.logit('default_yes ~ balance + income + student_yes', df).fit()
```

```python
print(m1_logit.summary2())
```

```python
# Valores que NO cumplen con el p-value en 95%
m1_logit.pvalues > 0.05
```
